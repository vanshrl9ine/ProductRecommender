```python
# Define a function to extract "Ideal For" value
def extract_ideal_for(spec):
    ideal_for_match = re.search(r'"key"=>"Ideal For", "value"=>"(.*?)"', spec)
    if ideal_for_match:
        return ideal_for_match.group(1)
    return None

# Apply the extraction function to create the "Ideal For" column
dataset['ideal_for'] = dataset['product_specifications'].apply(extract_ideal_for)

# Drop the original "product_specification" column if needed
dataset.drop(columns=['product_specifications'], inplace=True)


```

```python
dataset['ideal_for'].value_counts()
```

```python
def preprocess_ideal_for(value):
    if isinstance(value, str):  # Check if the value is a string
        value = re.sub(r"women's|women", "Women", value, flags=re.IGNORECASE)
        value = re.sub(r"men's|men", "men", value, flags=re.IGNORECASE)
        
        value = re.sub(r"Women's|Women", "Women", value, flags=re.IGNORECASE)
        value = re.sub(r"Men's|Men", "men", value, flags=re.IGNORECASE)
        
        
        
        value = re.sub(r"girls's|girls", "Girls", value, flags=re.IGNORECASE)
        value = re.sub(r"boy's|boys", "Boys", value, flags=re.IGNORECASE)
        
        value = re.sub(r"Girls's|Girls", "Girls", value, flags=re.IGNORECASE)
        value = re.sub(r"Boy's|Boys", "Boys", value, flags=re.IGNORECASE)
        
        
    return value

dataset['ideal_for'] = dataset['ideal_for'].apply(preprocess_ideal_for)

# Convert all values in "Ideal For" column to strings
dataset['ideal_for'] = dataset['ideal_for'].astype(str)
```

```python
indices = pd.Series(dataset.index,index=dataset['product_url']).drop_duplicates()
```

```python
def product_recommendation(title,sig=sig):
    indx = indices[title]
    
    
    sig_scores = list(enumerate(sig[indx]))
    
    #sorting products
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)
    
   
    sig_scores = sig_scores[1:11]
    
    
    product_indices = [i[0] for i in sig_scores]
    
    
    return dataset['product_url'].iloc[product_indices]
```
